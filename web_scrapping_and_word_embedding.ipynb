```python
#Scrapping the data from the website 
```

```python
import bs4 as bs
import urllib.request 
import requests
```

```python
scrap_data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Deep_learning')   #to load data
```

```python
article = scrap_data.read()
```

```python
parsed_article = bs.BeautifulSoup(article)   #to convert into a form that is understandable
```

```python
parsed_article
```

```python
parsed_article.title.string
```

```python
parsed_article.p   #gives only one paragraph
```

```python
paragraph = parsed_article.find_all('p')   #to get all the paragraphs and it only searches for html syntax p and not words
```

```python
paragraph
```

```python
for p in paragraph:
  print(p.text)
```

```python
article_text = ''
for p in paragraph:
  article_text = article_text + p.text
```

```python
article_text
```

```python
len(article_text)
```

```python
#Data cleaning
```

```python
import nltk
nltk.download('punkt')   #this is imported after running and to resolve the error
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')

import re
import pandas as pd
import numpy as np
```

```python
processed_article = article_text.lower()
```

```python
processed_article
```

```python
processed_article = re.sub('[^a-z]',' ',processed_article)
```

```python
processed_article
```

```python
all_sentence = nltk.sent_tokenize(processed_article)
```

```python
all_words = nltk.word_tokenize(processed_article)
```

```python
all_words_1 = [nltk.word_tokenize(sent)for sent in all_sentence]
```

```python
len(all_words_1)
```

```python
clean_text = [w for w in all_words_1 if w not in stop]
```

```python
clean_text
```

```python
len(clean_text)
```

```python
#Implementing word2vec
```

```python
from gensim.models import Word2Vec
```

```python
window_length = 3
embedding_dimension = 300
min_count = 1
no_of_cores = 4
epoch = 10
skipgram = 1  
```

```python
word2vec_1 = Word2Vec(clean_text,size = embedding_dimension,window=window_length,min_count=min_count,workers=no_of_cores,iter=epoch,sg=skipgram)
```

```python
vocabulary = word2vec_1.wv.vectors
```

```python
vocabulary
```

```python
word2vec_1.wv['learning']
```

```python
similar_words = word2vec_1.most_similar('learning',topn = 5)
```

```python
similar_words
```

```python
word2vec_1.most_similar(positive=['deep','neural'],negative=['learning'],topn=3)
```

```python
word2vec_1.save('skipgram.bin')
```

```python
skipgram_model = Word2Vec.load('skipgram.bin')
```
